```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
c0 = [1.0, 0.5]
c1 = [0.4, 0.8]
c02, c03 = c0
c12, c13 = c1
coordinates_lists1 = []
for order in range(21):
    P2 = np.exp(c02 + c12 * order)/(1 + np.exp(c02 + c12 * order) + np.exp(c03 + c13 * order))
    P3 = np.exp(c03 + c13 * order)/(1 + np.exp(c02 + c12 * order) + np.exp(c03 + c13 * order))
    P1 = 1 - P2 - P3
    coordinates_lists1.append([P1, P2, P3])

c1 = [0.2, 0.4]
c12, c13 = c1
coordinates_lists2 = []
for order in range(21):
    P2 = np.exp(c02 + c12 * order)/(1 + np.exp(c02 + c12 * order) + np.exp(c03 + c13 * order))
    P3 = np.exp(c03 + c13 * order)/(1 + np.exp(c02 + c12 * order) + np.exp(c03 + c13 * order))
    P1 = 1 - P2 - P3
    coordinates_lists2.append([P1, P2, P3])
    
c1 = [0.0, 0.0]
c12, c13 = c1
coordinates_lists3 = []
for order in range(21):
    P2 = np.exp(c02 + c12 * order)/(1 + np.exp(c02 + c12 * order) + np.exp(c03 + c13 * order))
    P3 = np.exp(c03 + c13 * order)/(1 + np.exp(c02 + c12 * order) + np.exp(c03 + c13 * order))
    P1 = 1 - P2 - P3
    coordinates_lists3.append([P1, P2, P3])
    
c1 = [-0.1, -0.05]
c12, c13 = c1
coordinates_lists4 = []
for order in range(21):
    P2 = np.exp(c02 + c12 * order)/(1 + np.exp(c02 + c12 * order) + np.exp(c03 + c13 * order))
    P3 = np.exp(c03 + c13 * order)/(1 + np.exp(c02 + c12 * order) + np.exp(c03 + c13 * order))
    P1 = 1 - P2 - P3
    coordinates_lists4.append([P1, P2, P3])
    
c1 = [-0.2, -0.1]
c12, c13 = c1
coordinates_lists5 = []
for order in range(21):
    P2 = np.exp(c02 + c12 * order)/(1 + np.exp(c02 + c12 * order) + np.exp(c03 + c13 * order))
    P3 = np.exp(c03 + c13 * order)/(1 + np.exp(c02 + c12 * order) + np.exp(c03 + c13 * order))
    P1 = 1 - P2 - P3
    coordinates_lists5.append([P1, P2, P3])
    
c1 = [-0.4, -0.8]
c12, c13 = c1
coordinates_lists6 = []
for order in range(21):
    P2 = np.exp(c02 + c12 * order)/(1 + np.exp(c02 + c12 * order) + np.exp(c03 + c13 * order))
    P3 = np.exp(c03 + c13 * order)/(1 + np.exp(c02 + c12 * order) + np.exp(c03 + c13 * order))
    P1 = 1 - P2 - P3
    coordinates_lists6.append([P1, P2, P3])
    
```

## Order size of $z=0$

```python
index = 0

X = [1, 2, 3]
x = np.arange(len(X))
width = 0.1

y1 = coordinates_lists1[index]
y2 = coordinates_lists2[index]
y3 = coordinates_lists3[index]
y4 = coordinates_lists4[index]
y5 = coordinates_lists5[index]
y6 = coordinates_lists6[index]

fig, ax = plt.subplots(figsize=(10, 6), dpi=100)
plt.rcParams['font.size'] = '16'

ax.grid(False)
plt.bar(x- 0.25, y1, width,  color='azure', edgecolor = 'black', label=f'(1, 0.5, 0.4, 0.8), z = {index}')
plt.bar(x- 0.15, y2, width,  color='skyblue', edgecolor = 'black', label=f'(1, 0.5, 0.2, 0.4), z = {index}')
plt.bar(x- 0.05, y3, width, color='royalblue', edgecolor = 'black', label=f'(1, 0.5, 0, 0), z = {index}')
plt.bar(x+ 0.05, y4, width, color='blue', edgecolor = 'black', label=f'(1, 0.5, -0.1, -0.05), z = {index}')
plt.bar(x+ 0.15, y5, width, color='mediumblue', edgecolor = 'black', label=f'(1, 0.5, -0.2, -0.1), z = {index}')
plt.bar(x+ 0.25, y6, width, color='darkblue', edgecolor = 'black', label=f'(1, 0.5, -0.4, -0.8), z = {index}')

plt.ylim(0,1)
plt.xticks(x,['1', '2', '3'], fontsize=20)
plt.yticks(fontsize=20)        
plt.xlabel("Remaining shelf-life",fontsize=25)
plt.ylabel("Probability",fontsize=25)
plt.legend(loc="upper center")
# remove spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
# grid
ax.set_axisbelow(True)
ax.yaxis.grid(color='gray', linestyle='dashed', alpha=0.7)

# plt.savefig('Endog_m3_Order0_v4.pdf', bbox_inches = 'tight') 
```
